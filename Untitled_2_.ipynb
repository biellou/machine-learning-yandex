```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

генерирую из разпределения парето выборку объёма 1000, строю гистограмму выборки и рисую поверх неё теоретическую плотность распределения случайной величины

```python
b = 2.62 #pareto принимает b как параметр формы
my_rv = sts.pareto(b)
sample = my_rv.rvs(1000)

x = np.linspace(0,5,100)
pdf = my_rv.pdf(x)
# теоретическая плотность распределения случайной величины
plt.plot(x, pdf, label='theoretical pdf')
# гистограмма выборки
plt.hist(sample, normed=True, label='histogram', bins = 50)
# метки
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

```python
#def gen_sample(n, sample): # генерация выборки выборочных средних для выборки объемом n
#    l = []
#    for i in xrange(1000):
#        np.random.shuffle(sample)
#        l.append(np.mean(sample[0:n]))
#    return l

def gen_sample(n, sample): # генерация выборки выборочных средних для выборки объемом n
    l = []
    sample = []
    for i in range(1,1000):
        sample.append(my_rv.rvs(n))
    for i in sample:
        l.append(np.mean(i))
    return l

var = my_rv.var() # вычисление дисперсии случайной величины для использования этого в предельной теореме
mean = my_rv.mean() # вычисление теоретического среднего случайной величины для использования этого в предельной теореме
print "теор. среднее = {0}, дисперсия = {1}".format(mean, var)
```

гистограмма и плотность соответствующего нормального распределения  для выборки с n = 5

```python
n = 5

# вычисление параметров для нормально распределения
mu = mean
sigma = (var/n)**0.5

x = np.linspace(0,5,100)
first_sample = gen_sample(n, sample) # генерация выборки с n = 5

plt.plot(x, sts.norm.pdf(x, loc=mu, scale=sigma), label='pdf')
plt.hist(first_sample, normed=True, label='hist')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

данная гистограмма слабо похожа на плотность нормального распределения

гистограмма и плотность соответствующего нормального распределения для выборки с n = 10

```python
n = 10

# вычисление параметров для нормально распределения
mu = mean
sigma = (var/n)**0.5

x = np.linspace(0,4,100)
second_sample = gen_sample(n, sample) # генерация выборки с n = 10
plt.plot(x, sts.norm.pdf(x,loc=mu, scale=sigma), label='pdf')
plt.hist(second_sample, normed=True, label='hist')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

данна гистограмма более похожа на плотность нормального распределения, чем предыдущее(т.к основная часть значений перешла от левого края ближе к центру), однако, напоминает нормальное распределение лишь отдаленно

гистограмма и плотность соответствующего нормального распределения для выборки с n = 50

```python
n = 50

# вычисление параметров для нормально распределения
mu = mean
sigma = (var/n)**0.5

x = np.linspace(0,4,100)
third_sample = gen_sample(n, sample) # генерация выборки с n = 50
plt.plot(x, sts.norm.pdf(x, loc=mu, scale=sigma), label='pdf')
plt.hist(third_sample, normed=True, label='hist')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

эта гистогограмма больше всех похожа на нормальное распределение, а также напоминает его, так как риснок напоминает шляпу

из проделанной работы можно видеть, что при n = 5 и n = 10 гистограмма лишь отдаленно напоминает нормальное распределение, а при n = 50 ассоциации уже появляются быстро. действительно, при росте n наша гистограмма все больше похожа на нормальное распределение.
